```python
import pandas as pd
import matplotlib.pyplot as plt 
import warnings
import dfview # https://github.com/xslyr/DataOverview

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

## Load, Shape e Head para simples conferência

```python
df = pd.read_csv('General_Files/train.csv')
df = df.drop(['Id'],axis=1)
print('\n\nDataframe de Treinamento: train.csv ({} linhas e {} colunas)\n'.format(df.shape[0],df.shape[1]))
df.head()
```


## Utilizando um describe personalizado p/ visualizar os tipos das colunas

PS: O describe padrão presente no pandas fornece dados apenas das colunas numéricas.

```python
dfinfo = dfview.DataOverview(df)
print('\nDistribuição dos tipos das colunas do dataframe.\n')
for i in dfinfo.columns_type():
    print('  {} colunas do tipo {}'.format(dfinfo.columns_type()[i], i))
```

### Describe das Colunas Tipo Int

```python
int_columns = dfinfo.show('int64')
int_columns.iloc[1:,:]
```

Observamos através do dataset acima apresentado que:
  - Nenhuma das colunas possuem valores nulos
  - Acompanhando as linhas min, mean, max e as informações de desvio padrão (std e std%) logo abaixo, conseguimos inferir mais ou menos onde temos outliers. Apartir dessas informações, juntamente com o dicionario de dados e o gráfico da distribuição dos dados, iremos escolher como tratar a coluna.
  
  
  Para conferir os conceitos e as diferenças entre Normalizar e Padronizar os dados, confira o link a seguir: https://medium.com/data-hackers/normalizar-ou-padronizar-as-vari%C3%A1veis-3b619876ccc9
  

```python
cols = list(int_columns.columns)
cols.remove('SalePrice')
d = df.sort_values('SalePrice',ascending=True).copy()

for col in cols:
    plt.scatter(d[col],d['SalePrice'])
    plt.title('{}'.format( dict(int_columns[col].iloc[1:-5]) ))
    plt.xlabel(col, fontsize=15)
    plt.ylabel('SalePrice')
    plt.show()
    print('\n\n')
```

Considerando a distribuição nos gráficos acima junto ao dicionário de dados, definimos as seguintes ações:

```python
cols2_categorize = ['MSSubClass','OverallQual','OverallCond']
cols2_standardize = ['YearBuilt','YearRemodAdd','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','HalfBath','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea']
cols2_normalize = ['LotArea','BsmtFinSF1','BsmtFullBath','BsmtHalfBath','FullBath','BedroomAbvGr''KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageCars','MiscVal','MoSold','YrSold']

print('\n\nConsiderando a distribuição nos gráficos acima junto ao dicionário de dados, definimos as seguintes ações:\n')
print('\tColunas a Padronizar:')
for i in cols2_standardize:
    print('\t\t{}'.format(i))
    
print('\n\t Colunas a Normalizar:')
for i in cols2_normalize:
    print('\t\t{}'.format(i))
    
print('\n\t Colunas a transformar em variáveis Categóricas:')
for i in cols2_categorize:
    print('\t\t{}'.format(i))
    
```

### Describe das Colunas Tipo Float

```python
float_columns = dfinfo.show('float64')
float_columns.iloc[1:,:]
```

Observamos através do dataset acima apresentado que:
  - A coluna LotFrontage contém ~18% de dados nulos
  - Provavelmente a coluna LotFrontage funcionará melhor com a Padronização enquanto GarageYrBlt basta uma Normalização.
  - Já a coluna MasVnrArea está com uma % de desvio padrão em 11,3%, portanto em nosso modelo inicial adotaremos também a Padronização.

```python
cols = list(float_columns.columns)
d = df.sort_values('SalePrice',ascending=True).copy()

for col in cols:
    plt.scatter(d[col],d['SalePrice'])
    plt.title('{}'.format( dict(float_columns[col].iloc[1:-5]) ))
    plt.xlabel(col, fontsize=15)
    plt.ylabel('SalePrice')
    plt.show()
    print('\n\n')
```

```python
print('Nos 3 casos iremos incluí-las na lista de colunas a serem padronizadas.')

cols2_standardize += list(float_columns)
```
